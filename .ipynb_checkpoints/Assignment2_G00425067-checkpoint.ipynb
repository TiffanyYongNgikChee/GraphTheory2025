```python
# From assignment 1 -method, to return the adjacent dictionary
def returnAdjacentVertices(vertices, edges):
    
    adjVertices = {v: [] for v in vertices}  # Initialize an adjacency list dictionary with empty lists for each vertex
    
    # Iterate through each vertex in the adjacency list
    for vertex1, vertex2 in edges:
        adjVertices[vertex1].append(vertex2) # Add vertex2 as a neighbor of vertex1
        adjVertices[vertex2].append(vertex1) # Add vertex1 as a neighbor of vertex2 (undirected graph)
    
    return adjVertices # Return the adjacency list representation
```

```python
# From assignment 1 - using the adjacent dictionary and made it to number(degree)
def returnDegreesOfAdjacentVertices(adjVertices):
    
    degreeDict = {} # Initialize an empty dictionary to store results
    
    # Iterate through each vertex and its list of adjacent vertices
    for vertex, adjacents in adjVertices.items():
        # Compute the degrees of all adjacent vertices
        degrees = [len(adjVertices[adjacent]) for adjacent in adjacents]
        # Store the sorted list of degrees in the dictionary
        degreeDict[vertex] = sorted(degrees) 
    
    return degreeDict # Return the final dictionary
```

```python
# Method 1 - takes in two dictionaries, and outputs another dictionary saying which vertices in Graph 2 can be mapped to Graph 1.
def returnPossibleMappingsDicionary(adj1,adj2):
    # Initialize a dictionary to hold possible matches from adj1 to adj2
    possibleMappings = {}
    # For each vertex in the first graph's degree dictionary
    for d in adj1:
        # Initialize an empty set to store matching vertices from the second graph
        possibleMappings[d] = set()
        # Compare this vertex's degree signature with every vertex in the second graph
        # If the degree signature matches, consider it a potential mapping
        for v in adj2:
            if adj1[d] == adj2[v]:
                possibleMappings[d].add(v)
    # Return the dictionary of all possible vertex matches based on degree signature
    return possibleMappings
```

```python
# Method 1 - Using itertools product, to generate all combinations of mappings based on the sets in possibleMappings.
def returnPossibleMappings(possMap):
    # Initialize a list to hold all combinations of possible mappings
    listPossibleMappings = []
    # Extract the list of possible values (sets of possible matches) from the dictionary
    values = list(possMap.values())
    # Compute the Cartesian product of the sets, resulting in tuples of all possible mappings
    return list(product(*values)) # Return all combinations of mappings as tuples
```

```python
# Method 2 - in particular which mappings are isomorphisms.
def createChoice(mydictionary):
    mappings = [[]] # Start with an empty mapping list (initial state of decision tree)
    
    for d in mydictionary: # For each vertex in the first graph
        current_mappings = [] # Temporary list to store valid extensions of current mappings
        choices = mydictionary[d] # Get possible target vertices for current vertex
        for m in mappings: # For each existing mapping path
            for c in choices: # For each potential match
                if c not in m: # Ensure the mapping remains injective (no duplicates)
                    current_mappings.append(m+[c]) # Extend the current mapping
        
        mappings = current_mappings # Move to next level of decision tree with extended mappings
    
    return mappings # Return list of all valid injective mappings
```

```python
# Method 2 - translates the edges of the first graph using the labels of the second graph
def translateEdges(edges1,vertices1,map1):
    
    count = 0  # Initialize index counter for mapping
    mapDict = {} # Dictionary to store vertex mapping from graph1 to graph2
    
    edges1_translated = []  # List to hold translated edges
    
    for v in vertices1: # Loop through each vertex in graph1
        mapDict[v] = map1[count] # Map vertex from graph1 to graph2 using index
        count = count + 1 # Increment index
        
    for e in edges1: # Loop through each edge in graph1
        u,v = e # Unpack the edge
        edges1_translated.append({mapDict[u],mapDict[v]}) # Create a set representing the translated edge (undirected)
        
    return edges1_translated # Return translated edges based on current mapping
```

```python
# Method 2 - Compare to see if translated edges match the second graph.
def checkEdges(newEdges, edges2):
    # Convert edges in graph2 into sets for easy comparison (undirected)
    edge2Set = [frozenset(e) for e in edges2]
    
    for e in newEdges: # Check each translated edge
        if frozenset(e) not in edge2Set: # If any edge is not in graph2
            return False  # If any edge is missing, not an isomorphism
    
    # Ensure all edges in graph2 are present in translated set
    newEdgeSet = [frozenset(e) for e in newEdges]
    for e in edge2Set:
        if e not in newEdgeSet:
            return False  # Edge from graph2 missing in translated graph

    return True  # All edges match: graphs are isomorphic under this mapping
```

```python
# Using method 2 for test cases
def findIsomorphismsMethod_2(vertices1, edges1, vertices2, edges2):
    # Step 1: Build adjacency list for both graphs
    adj1 = returnAdjacentVertices(vertices1, edges1)
    adj2 = returnAdjacentVertices(vertices2, edges2)

    # Step 2: Create degree pattern dictionaries for both graphs
    deg1 = returnDegreesOfAdjacentVertices(adj1)
    deg2 = returnDegreesOfAdjacentVertices(adj2)

    # Step 3: Determine possible vertex mappings based on degree patterns
    possibleMappings = returnPossibleMappingsDicionary(deg1, deg2)

    # Step 4: Generate all injective (valid) combinations using decision tree
    allMappings = createChoice(possibleMappings)

    # Step 5: Test each candidate mapping
    isomorphisms = [] # List to store all valid isomorphism mappings
    for mapping in allMappings:
        # Translate edges of graph1 using the current mapping
        translatedEdges = translateEdges(edges1, vertices1, mapping)
        # Check if translated edges match edges of graph2
        if checkEdges(translatedEdges, edges2):
            # If they match, record the mapping as a valid isomorphism
            iso_map = dict(zip(vertices1, mapping)) # Convert list to dict for readability
            isomorphisms.append(iso_map)

    return isomorphisms # Return all isomorphic mappings
```

```python
# Method 3 - implements the adjacency matrix isomorphism check using : 𝐴1 = 𝑃𝐴2𝑃𝑇
# Converts a list of vertices and edges into an adjacency matrix.
def returnAdjacentMatrix(vertices,edges):
    # Initialize an N x N matrix of zeros (N = number of vertices)
    M = np.zeros([len(vertices),len(vertices)],dtype = int)
    
    for e in edges: # Loop through each edge
        coordinates = [] # Store indices of the two vertices
        for v in e:
            coordinates.append(vertices.index(v)) # Find vertex index in the vertex list
        # Set matrix entries to 1 for undirected edge (i,j) and (j,i)
        M[coordinates[0]][coordinates[1]] = 1
        M[coordinates[1]][coordinates[0]] = 1
        
    return M # Return the adjacency matrix
```

```python
# Method 3 - Returns the transpose of a given matrix
def transpose(A): 
    # Create an empty transposed matrix with flipped dimensions
    At = np.zeros([len(A[0]),len(A)], dtype=int)
    for i in range(len(A[0])): # Columns of A become rows of At
        for j in range(len(A)): # Rows of A become columns of At
            At[i][j] = A[j][i] # Transpose element assignment
    return At # Return transposed matrix
```

```python
# Method 3 - Generates all permutation matrices P, and checks which satisfy the condition A1=PA2PT to find graph isomorphisms 
def findMatrixIsomorphisms(vertices1, edges1, vertices2, edges2):
    # Step 1: Compute adjacency matrices for both graphs
    M1 = returnAdjacentMatrix(vertices1, edges1)
    M2 = returnAdjacentMatrix(vertices2, edges2)

    n = len(vertices1) # Number of vertices in the graphs
    identity = np.identity(n, dtype=int) # Identity matrix (not directly used here)
    perm_indices = permutations(range(n)) # Generate all permutations of vertex indices

    valid_permutations = [] # List to store valid isomorphism permutations
    for perm in perm_indices:
        # Step 2: Build permutation matrix P for this permutation
        P = np.zeros((n, n), dtype=int)
        for i in range(n):
            P[i, perm[i]] = 1 # Place 1 in each row according to the permutation
        
        # Step 3: Apply the transformation: P A2 Pᵀ
        A2_transformed = np.dot(P, np.dot(M2, transpose(P)))
        # Step 4: Check if the transformed A2 equals A1
        if np.array_equal(M1, A2_transformed):
            valid_permutations.append((P, perm))  # Store matrix and corresponding permutation

    return valid_permutations # Return all isomorphisms found
```

```python
# Method 4 - Efficient Isomorphism Detection via Filtered Permutation Matrices
# Converts a mapping of vertices into a permutation matrix P
def convertMappingToPermutationMatrix(mapping, vertices):
    """
    Convert a mapping like ['s', 'p', 'r', 'u', 'q', 't'] into a permutation matrix P,
    assuming original vertex order is like ['p', 'q', 'r', 's', 't', 'u']
    """
    n = len(vertices)
    P = np.zeros((n, n), dtype=int) # Create an empty n x n matrix
    for i, mapped_vertex in enumerate(mapping):
        j = vertices.index(mapped_vertex) # Find index of mapped vertex in original list
        P[i][j] = 1 # Place 1 in the permutation matrix at (i, j)
    return P # Return the constructed permutation matrix
```

```python
# Method 4 - Takes all possible mappings and builds permutation matrices using Graph 2's vertex order.
def buildPermutationMatricesFromMappings(mappings, vertices2):
    """
    Convert all possible vertex mappings into permutation matrices
    using the known vertex ordering of Graph 2.
    """
    matrices = []
    for m in mappings:
        P = convertMappingToPermutationMatrix(m, vertices2) # Build P for each mapping
        matrices.append((P, m)) # Store both the matrix and the mapping
    return matrices # Return list of all (P, mapping) pairs
```

```python
# Method 4 - Takes in adjacency matrices and checks each permutation matrix to verify isomorphism.
def checkIsomorphicPermutationMatrices(M1, M2, permutationMatrices):
    """
    Given adjacency matrices and P matrices, return all valid isomorphisms.
    """
    valid = []
    for P, mapping in permutationMatrices:
        A2_transformed = np.dot(P, np.dot(M2, transpose(P))) # Compute P A2 Pᵗ
        if np.array_equal(M1, A2_transformed): # If result equals A1, it's a valid isomorphism
            valid.append((P, mapping))
    return valid # Return only the correct isomorphisms
```

```python
# Method 4 - Main function for Method 4: Filters using Method 1, then checks using matrix isomorphism logic.
def findIsomorphismsWithMatrixFiltering(vertices1, edges1, vertices2, edges2):
    # Step 1: Compute adjacency matrices
    M1 = returnAdjacentMatrix(vertices1, edges1)
    M2 = returnAdjacentMatrix(vertices2, edges2)

    # Step 2: Generate degree-based mapping dictionary
    adj1 = returnAdjacentVertices(vertices1, edges1)
    adj2 = returnAdjacentVertices(vertices2, edges2)

    deg1 = returnDegreesOfAdjacentVertices(adj1)
    deg2 = returnDegreesOfAdjacentVertices(adj2)
    # Use degree pattern to filter potential mappings
    possibleMappingsDict = returnPossibleMappingsDicionary(deg1, deg2)
    filteredMappings = createChoice(possibleMappingsDict) # Fewer mappings than brute-force

    # Step 3: Convert filtered mappings to permutation matrices
    P_matrices = buildPermutationMatricesFromMappings(filteredMappings, vertices2)

     # Step 4: Check which permutation matrices are actual isomorphisms
    isomorphisms = checkIsomorphicPermutationMatrices(M1, M2, P_matrices)

    return isomorphisms # Return all valid isomorphisms
```
